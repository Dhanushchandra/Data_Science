```python
import wordcloud
import nltk
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.probability import FreqDist
```

```python
data = open("./data/novel.txt","r").read()
```

```python
data
```

```python
tokens = word_tokenize(data)
```

```python
tokens
```

```python
nltk.download("punkt")
```

```python
stopwords_list = stopwords.words("english")
```

```python
stopwords_list
```

```python
no_punc = []
for word in tokens:
    if word.isalpha():
        no_punc.append(word.lower())
no_punc
```

```python
clean_words=[]
for word in no_punc:
    if word not in stopwords_list:
        clean_words.append(word)
```

```python
clean_words
```

```python
main_words=" ".join(clean_words)
main_words
```

```python
fq = FreqDist(clean_words)
fq
```

```python
wordcloud = WordCloud(width=800,height=400, background_color="white").generate_from_frequencies(fq)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
```
